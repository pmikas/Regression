Functions

```python
#   Compute the hypothesis function
def hypothesis(theta,X):
    h = np.matmul(X,theta)
    return h

#   Compute the cost by Gradient Descent
def costGradientDescent(X, Y, alpha, num_it):
    m,n = np.shape(X)
    theta = np.zeros(n)
    J = np.zeros(num_it)
    h = hypothesis(theta,X)
    xt = np.transpose(X)

    for i in range(0,num_it):
        theta = theta - alpha/m*np.matmul(xt,h-Y)    
        h = hypothesis(theta,X)
        J[i] = 1/(2*m)*np.matmul((h-Y),(h-Y))
    return J

#   Compute the cost by Normal Equation
def costNormalEquation(X, Y):
    m,n = np.shape(X)
    J = 0
    theta = NormalEquation(X,Y)
    h = hypothesis(theta,X)
    J = 1/(2*m)*np.matmul((h-Y),(h-Y))
    return J

def NormalEquation(X,Y):
    m,n = np.shape(X)
    theta = np.zeros(n)  
    xt = np.transpose(X)
    theta = np.matmul(np.matmul(np.linalg.inv(np.matmul(xt,X)),xt),Y)
    return theta

#   Compute the feature scaling
def FeatureScaling(feat):
    new_feat = feat/np.max(feat)
    return new_feat

#   Compute the mean normalization
def MeanNormalization(feat):
    mean_feat = np.mean(feat)
    new_feat = (feat-mean_feat)/np.std(feat)
    return new_feat
```

```python
# Load the file "Diamonds"
data = pd.read_csv("diamonds.csv")
data.head()
```

```python
# Mapping the categorical features

#   a: x0
data = data.rename(index=str,columns={"Unnamed: 0": "a"})
data.a = 1

# Cut: quality of the diamond. Quality in increasing order Fair, Good, Very Good, Premium, Ideal
map_cut = {'Fair': 1, 'Good': 2, 'Very Good': 3, 'Premium': 4, 'Ideal': 5}
data.cut = data.cut.replace(map_cut)

# Color: D being the best and J the worst
map_color = {'J': 1, 'I': 2, 'H': 3, 'G': 4, 'F': 5, 'E':6 ,'D':7}
data.color = data.color.replace(map_color)

#  Clarity: measurement of how clear the diamond is (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))
map_clarity = {'I1': 1, 'SI2': 2, 'SI1': 3, 'VS2': 4, 'VS1': 5, 'VVS2': 6, 'VVS1': 7, 'IF': 8}
data.clarity = data.clarity.replace(map_clarity)
```

```python
train = 45849   # Number of training examples
#cv =    # Number of cross-validation examples
test = 8091    # Number of test examples

data_train = data[:train]
#data_cv = data_train[:]
data_test = data[:test]

Xtrain = data_train.drop(['price'], axis=1)
Ytrain = data_train.price

#Xcv = 
#Ycv = 

Ytest = data_test.price
Xtest = data_test.drop(['price'], axis=1)
```

```python
# Remove outliers from dimensions x, y and z
data_train = data_train[(data_train[['x','y','z']] != 0).all(axis=1)]
data_train = data_train[(data_train[['y','z']] < 30).all(axis=1)]
```

```python
# Visualization of the relation between the features
#sns.set()
cols = ['price','carat','cut','color','clarity','depth','table','x','y','z']
#sns.pairplot(data[cols])
#plt.show
```

```python
# Correlation between the features and the price
data_train[cols].corr()
```

```python
# Feature scaling and rearranging the data_train

#   Carat: weight of the diamond
data_train.carat = FeatureScaling(data_train.carat)

#   Cut: quality of the diamond
data_train.cut = FeatureScaling(data_train.cut)

#   Color
data_train.color = FeatureScaling(data_train.color)

#   Clarity: measurement of how clear the diamond is 
data_train.clarity = FeatureScaling(data_train.clarity)

#   Depth: total depth percentage = z / mean(x, y) = 2 * z / (x + y)
data_train.depth = FeatureScaling(data_train.depth)

#   Table: width of top of diamond relative to widest point
data_train.table = FeatureScaling(data_train.table)

#   x in mm
data_train.x = FeatureScaling(data_train.x)

#   y in mm
data_train.y = FeatureScaling(data_train.y)

#   z in mm
data_train.z = FeatureScaling(data_train.z)
```

```python
Xtrain.x = np.power(Xtrain.x,4)
Xtrain.y = np.power(Xtrain.y,4)
Xtrain.z = np.power(Xtrain.z,3)
```

```python
alpha = [0.01,0.03,0.1,0.3,1,3,10,30]    # learning rate
num_it = 1000   # Number of iterations to find theta
```

```python
# Compute the cost using the training examples
JNE = costNormalEquation(Xtrain,Ytrain)                     # call the Normal Equation function
CLF = linear_model.SGDRegressor()
CLF.fit(Xtrain,Ytrain)
JSL = CLF.predict(Xtest)
print('Cost of the Normal Equation:', JNE)
print('Cost of the Scikit-Learn:', np.min(JSL))
#plt.plot(range(0,np.size(JSL)),JSL)
```

```python
JGD = costGradientDescent(Xtrain,Ytrain,alpha[0],num_it)    # call the Gradient Descent function
plt.plot(range(0,num_it),JGD)
plt.xlabel('Number of iterations')
plt.ylabel('Cost (J)')
print('Cost of the Gradient Descent:', np.min(JGD))
```
